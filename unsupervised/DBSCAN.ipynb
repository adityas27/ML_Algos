```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
class DBSCAN:

    def __init__(self, eps=0.8, min_samples=5):
        self.eps = eps
        self.min_samples = min_samples

    def region_query(self, x, point_idx):
        distances = np.linalg.norm(x - x[point_idx], axis=1)
        return np.where(distances <= self.eps)[0]

    def expand_cluster(self, x, labels, point_idx, neighbors, cluster_id, visited):
        labels[point_idx] = cluster_id

        i = 0

        while i < len(neighbors):

            neighbor_idx = neighbors[i]

            if not visited[neighbor_idx]:
                visited[neighbor_idx] = True

                new_neighbors = self.region_query(x, neighbor_idx)

                if len(new_neighbors) >= self.min_samples:
                    neighbors = np.concatenate((neighbors, new_neighbors))

            if labels[neighbor_idx] == 0:
                labels[neighbor_idx] = cluster_id

            i += 1
```

```python
def fit(self, x):
    n_samples = len(x)

    labels = np.zeros(n_samples, dtype=int)

    visited = np.zeros(n_samples, dtype=bool)

    cluster_id = 0

    for point_idx in range(n_samples):

        if visited[point_idx]:
            continue

        visited[point_idx] = True

        neighbors = self.region_query(x, point_idx)

        if len(neighbors) < self.min_samples:
            labels[point_idx] = -1

        else:
            cluster_id += 1

            self.expand_cluster(
                x,
                labels,
                point_idx,
                neighbors,
                cluster_id,
                visited
            )

    self.labels_ = labels

DBSCAN.fit = fit
```

```python
def fit_predict(self, x):
    self.fit(x)
    return self.labels_
DBSCAN.fit_predict = fit_predict
```

```python
iris = load_iris()
x = iris.data
```

```python
model = DBSCAN(eps=0.8, min_samples=5)
labels = model.fit_predict(x)
```

```python
print("Unique Clusters:", np.unique(labels))

print("\nFirst 20 Labels:")
print(labels[:20])

print("\nCluster Counts:")
for cluster in np.unique(labels):
    print(cluster, np.sum(labels == cluster))
```
