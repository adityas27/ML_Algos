```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
class GaussianNB:

    def fit(self, x, y):
        self.classes = np.unique(y)

        self.mean = {}
        self.var = {}
        self.prior = {}

        for cls in self.classes:
            x_cls = x[y == cls]

            self.mean[cls] = np.mean(x_cls, axis=0)
            self.var[cls] = np.var(x_cls, axis=0) + 1e-9
            self.prior[cls] = len(x_cls) / len(x)
    def gaussian(self, cls, sample):
        mean = self.mean[cls]
        var = self.var[cls]

        numerator = np.exp(-((sample - mean) ** 2) / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)

        return numerator / denominator
```

```python
def predict(self, x):
    predictions = []

    for sample in x:

        scores = []
        for cls in self.classes:
            prior = np.log(self.prior[cls])
            likelihood = np.sum(np.log(self.gaussian(cls, sample)))
            scores.append(prior + likelihood)

        predictions.append(
            self.classes[np.argmax(scores)]
        )

    return np.array(predictions)

GaussianNB.predict = predict
```

```python
def predict_proba(self, x):
    probabilities = []
    for sample in x:
        scores = []
        for cls in self.classes:
            prior = np.log(self.prior[cls])
            likelihood = np.sum(
                np.log(
                    self.gaussian(cls, sample)
                )
            )

            scores.append(prior + likelihood)

        scores = np.exp(scores - np.max(scores))
        scores = scores / np.sum(scores)
        probabilities.append(scores)

    return np.array(probabilities)

GaussianNB.predict_proba = predict_proba
```

```python
data = load_breast_cancer()
```

```python
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = GaussianNB()

model.fit(x_train, y_train)
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test, predictions)
```

```python
print("Accuracy:", accuracy)

print("\nFirst 10 Predictions:")
print(predictions[:10])

print("\nProbabilities:")
print(model.predict_proba(x_test[:5]))
```
