```python
class LinearRegression():
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.W = None
        self.b = None
        self.losses = []
        self.mean = None
        self.std = None
        self.tolerance = 1e-6
    def predict(self, X):
        X = self.normalize(X)
        return X @ self.W + self.b

    def compute_loss(self, y, y_pred):
        m = len(y) # number of sample
        return (1 / m) * np.sum((y_pred - y) ** 2) # loss function (MSE: Minimum squared ereor)

    def compute_gradients(self, X, y, y_pred):

        m = len(y) # number of sample
        error = y_pred - y # this is matrix

        dW = (2 / m) * (X.T @ error)
        db = (2 / m) * np.sum(error)

        return dW, db

    def fit(self, X, y):

        X = np.array(X)
        y = np.array(y).reshape(-1, 1)

        m, n = X.shape

        self.W = np.zeros((n, 1)) # set to zeros
        # self.W = np.random.randn(0, 1, size=(n, 1)) # set with random numbers bw 0 to 1
        self.b = 0
        prev_loss = float("inf")
        for _ in range(self.iterations):
            y_pred = self.predict(X)

            current_loss = self.compute_loss(y, y_pred)
            if abs(prev_loss - current_loss) < self.tolerance:
                break        
            prev_loss = current_loss
            self.losses.append(current_loss)

            dW, db = self.compute_gradients(X, y, y_pred)

            self.W -= self.learning_rate * dW
            self.b -= self.learning_rate * db

    def get_parameters(self):
        return self.W, self.b
```

```python
# normalisation
def normalize_1F(self, X):
    mean = np.mean(X)
    std = np.std(X)
    for i in range(len(X)):
        X[i] = (X[i]-mean)/std
    return X

LinearRegression.normalize_1F = normalize_1F
    
def normalize(self, X):
    X = np.array(X, dtype=np.float64)

    if self.mean is None:
        self.mean = np.mean(X, axis=0)
    if self.std is None:
        self.std = np.std(X, axis=0)

    self.std[self.std == 0] = 1
    return (X - self.mean) / self.std

LinearRegression.normalize = normalize
```

```python
def r2_score(self, y, y_pred):
    y = np.array(y).reshape(-1, 1)
    y_pred = np.array(y_pred).reshape(-1, 1)

    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r2 = 1 - (ss_res / ss_tot) # r^2 formula

    return r2

LinearRegression.r2_score = r2_score
```

```python
def train_test_split(self, X, y, test_size=0.2):
    X = np.array(X)
    y = np.array(y)
    m = len(X)
    
    indices = np.random.permutation(m)
    test_count = int(m * test_size)
    
    test_indices = indices[:test_count]
    train_indices = indices[test_count:]
    
    X_train = X[train_indices]
    X_test = X[test_indices]

    y_train = y[train_indices]
    y_test = y[test_indices]

    return X_train, X_test, y_train, y_test

LinearRegression.train_test_split = train_test_split
```
