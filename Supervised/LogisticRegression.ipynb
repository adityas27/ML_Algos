```python
import numpy as np
```

```python
class LogisticRegression():
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.W = None
        self.b = None
        self.losses = []
        self.mean = None
        self.std = None
        self.tolerance = 1e-6
    def predict(self, X):
        X = self.normalize(X)
        z = X @ self.W + self.b
        probabilities = 1 / (1 + np.exp(-z)) # sigmoid implementation 
        return (probabilities >= 0.5).astype(int)
    def compute_loss(self, y, y_pred):
        m = len(y)
        epsilon = 1e-10
        loss = -(1 / m) * np.sum(y * np.log(y_pred + epsilon) + (1 - y) * np.log(1 - y_pred + epsilon))
        return loss
```

```python
def normalize(self, X):
    X = np.array(X, dtype=np.float64)

    if self.mean is None:
        self.mean = np.mean(X, axis=0)
    if self.std is None:
        self.std = np.std(X, axis=0)

    self.std[self.std == 0] = 1
    return (X - self.mean) / self.std

LogisticRegression.normalize = normalize
```

```python
def compute_gradients(self, X, y, y_pred):
    m = X.shape[0]

    # gradient for weights
    dW = (1 / m) * (X.T @ (y_pred - y))

    # gradient for bias
    db = (1 / m) * np.sum(y_pred - y)

    return dW, db

LogisticRegression.compute_gradients = compute_gradients
```

```python
def train_test_split(self, X, y, test_size=0.2):
    X = np.array(X)
    y = np.array(y)
    m = len(X)
    
    indices = np.random.permutation(m)
    test_count = int(m * test_size)
    
    test_indices = indices[:test_count]
    train_indices = indices[test_count:]
    
    X_train = X[train_indices]
    X_test = X[test_indices]

    y_train = y[train_indices]
    y_test = y[test_indices]

    return X_train, X_test, y_train, y_test

LogisticRegression.train_test_split = train_test_split
```

```python
def fit(self, X, y):
    X = np.array(X)
    y = np.array(y).reshape(-1, 1)

    m, n = X.shape

    self.W = np.zeros((n, 1)) # set to zeros
    # self.W = np.random.randn(0, 1, size=(n, 1)) # set with random numbers bw 0 to 1
    self.b = 0
    prev_loss = float("inf")
    for _ in range(self.iterations):
        y_pred = self.predict(X)

        current_loss = self.compute_loss(y, y_pred)
        if abs(prev_loss - current_loss) < self.tolerance:
            break        
        prev_loss = current_loss
        self.losses.append(current_loss)

        dW, db = self.compute_gradients(X, y, y_pred)

        self.W -= self.learning_rate * dW
        self.b -= self.learning_rate * db

def get_parameters(self):
    return self.W, self.b

LogisticRegression.fit = fit
LogisticRegression.get_parameters = get_parameters
```
